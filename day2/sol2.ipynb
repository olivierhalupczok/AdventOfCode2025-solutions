```python
def isInvalidId(x):
    if x[0] == "0":
        return False
    for k in range(1, len(x) // 2 + 1):
        if len(x) % k == 0 and x[:k] * (len(x) // k) == x:
            return True
    return False

```

```python
def return_invalid_ids(ranges):
    if not ranges:
        return []
    
    invalid_ids = []
    ranges_bounds = []
    
    # Handle both string and list inputs
    if isinstance(ranges, str):
        range_list = ranges.split(',') if ',' in ranges else [ranges]
    else:
        range_list = ranges
    
    for r in range_list:
        start, end = map(int, r.split('-'))
        ranges_bounds.append((start, end))
    
    for r in ranges_bounds:
        for i in range(r[0], r[1] + 1):
            if isInvalidId(str(i)):
                invalid_ids.append(i)
    
    return invalid_ids

# return_invalid_ids(ranges)
print(return_invalid_ids(ranges)[0])
sum(return_invalid_ids(ranges))
```

```python
def test(f, expected):
    return f == expected

def print_test(f, expected):
    result = f == expected
    print(result if result else f"Failed: {f} != {expected}")



```
