```python
with open("input4.txt") as f:
    lines = f.read().strip().split("\n")

def validate_arg(n, l):
    return n >= 0 and n < l

def get_adjacent_count(x, y, lenX, lenY):
    _count = 0
    for i in range(x-1, x+2):
        for j in range(y-1, y+2):
            if i == x and j == y:
                continue
            if validate_arg(i, lenX) and validate_arg(j, lenY):
                _count += (lines[i][j] == "@")
                # lines[i] = lines[i][:j] + 'X' + lines[i][j+1:]
                # print(lines[i])

    return _count

print(type(lines))
print(type(lines[0]))
print(type(lines[0][0]))
```

```python
count = 0
changed = True

while changed:
    changed = False
    to_remove = []
    for x in range(len(lines)):
        for y in range(len(lines[x])):
            if lines[x][y] == "@":
                # print(f"char: {lines[x][y]}, x: {x}, y: {y}, count: {get_adjacent_count(x, y, len(lines), len(lines[x]))}")
                isAccessible = (get_adjacent_count(x, y, len(lines), len(lines[x])) < 4)
                if isAccessible:
                    to_remove.append((x, y))
    if len(to_remove) > 0:
        changed = True  
        count += len(to_remove)
        for (x, y) in to_remove:
            lines[x] = lines[x][:y] + '.' + lines[x][y+1:]

count
```
